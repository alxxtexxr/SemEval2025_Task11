```python
import pandas as pd
```

```python
base_df = pd.read_csv('submissions/pred_sun_test_gpt4o_avg_5.csv')
# anger_df = pd.read_csv('submissions/track_a_42_1/pred_sun.csv')
disgust_df = pd.read_csv('submissions/pred_sun_test_disgust_63_3.csv')
fear_df = pd.read_csv('submissions/pred_sun_test_fear_42.csv')
surprise_df = pd.read_csv('submissions/pred_sun_test_surprise_42_1.csv')

id_df = base_df[['id']]
```

```python
# anger_df.loc[anger_df['anger'] != 1, :] = 0
disgust_df.loc[disgust_df['disgust'] != 1, :] = 0
fear_df.loc[fear_df['fear'] != 1, :] = 0
surprise_df.loc[surprise_df['surprise'] != 1, :] = 0
```

```python
dfs = [base_df, 
       # anger_df, 
       disgust_df, 
       fear_df, 
       surprise_df]
dfs = [df.drop('id', axis=1) for df in dfs]
```

```python
sum_df = sum(dfs)
sum_df
```

```python
one_hot_df = (sum_df >= 1).astype(int)
one_hot_df
```

```python
base_df.sum()
```

```python
disgust_df.sum()
```

```python
one_hot_df.sum()
```

```python
save_df = pd.concat([id_df, one_hot_df], axis=1)
save_df
```

```python
save_path = 'pred_sun_test_1.csv'
save_df.to_csv(save_path, index=False)
print("Saved to:", save_path)
```
