```python
import pandas as pd
```

```python
base_df_ = pd.read_csv('submissions/pred_sun_test_gpt4o_avg_10.csv')
# anger_df = pd.read_csv('submissions/track_a_42_1/pred_sun.csv')
disgust_df_ = pd.read_csv('submissions/pred_sun_test_disgust_63_3.csv')
fear_df_ = pd.read_csv('submissions/pred_sun_test_fear_42.csv')
surprise_df_ = pd.read_csv('submissions/pred_sun_test_surprise_42_1.csv')

id_df = base_df_[['id']]

base_df = base_df_.drop('id', axis=1)
disgust_df = disgust_df_.drop('id', axis=1)
fear_df = fear_df_.drop('id', axis=1)
surprise_df = surprise_df_.drop('id', axis=1)

emotion_cols = base_df.columns
whitelist_cols = ['disgust', 'surprise']
```

```python
# anger_df.loc[anger_df['anger'] != 1, :] = 0
```

```python
print("Disgust DF:")
print(disgust_df_[emotion_cols].sum())
print()

# disgust_df.loc[disgust_df['disgust'] != 1, :] = 0
disgust_df.loc[:, ~emotion_cols.isin(['disgust'] + whitelist_cols)] = 0

print("Disgust DF (after processed):")
print(disgust_df.sum())
```

```python
print("Fear DF:")
print(fear_df_[emotion_cols].sum())
print()

# fear_df.loc[fear_df['fear'] != 1, :] = 0
fear_df.loc[:, ~emotion_cols.isin(['fear'] + whitelist_cols)] = 0

print("Fear DF (after processed):")
print(fear_df.sum())
```

```python
print("Surprise DF:")
print(surprise_df_[emotion_cols].sum())
print()

# surprise_df.loc[surprise_df['surprise'] != 1, :] = 0
surprise_df.loc[:, ~emotion_cols.isin(['surprise'] + whitelist_cols)] = 0

print("Surprise DF (after filtered):")
print(surprise_df.sum())
```

```python
print("Base DF:")
print(base_df_[emotion_cols].sum())
print()

base_df.loc[:, ~emotion_cols.isin(['anger', 'joy', 'sadness'] + whitelist_cols)] = 0

print("Base DF (after filtered):")
print(base_df.sum())
```

```python
dfs = [base_df, 
       # anger_df, 
       disgust_df, 
       fear_df, 
       surprise_df]
```

```python
sum_df = sum(dfs)
sum_df
```

```python
one_hot_df = (sum_df >= 1).astype(int)
one_hot_df
```

```python
save_df = pd.concat([id_df, one_hot_df], axis=1)
save_df
```

```python
save_path = 'pred_sun_test_3.csv'
save_df.to_csv(save_path, index=False)
print("Saved to:", save_path)
```
