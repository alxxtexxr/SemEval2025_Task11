```python
import os
import pandas as pd
```

```python
submission_paths = [
    'submissions/pred_sun_test_gpt4o_1.csv',
    'submissions/pred_sun_test_gpt4o_2.csv',
    'submissions/pred_sun_test_gpt4o_3.csv',
    'submissions/pred_sun_test_gpt4o_4.csv',
    'submissions/pred_sun_test_gpt4o_5.csv',
]
dfs = [pd.read_csv(submission_path) for submission_path in submission_paths]
id_df = dfs[0][['id']]
dfs = [df.drop('id', axis=1) for df in dfs]
dfs[0]
```

```python
avg_df = sum(dfs)/len(dfs)
avg_df
```

```python
one_hot_df = (avg_df >= 0.5).astype(int)
one_hot_df
```

```python
dfs[0].sum()
```

```python
one_hot_df.sum()
```

```python
len(one_hot_df)
```

```python
save_df = pd.concat([id_df, one_hot_df], axis=1)
save_df
```

```python
save_path = 'pred_sun_test_gpt4o_avg_5.csv'
save_df.to_csv(save_path, index=False)
print("Saved to:", save_path)
```
