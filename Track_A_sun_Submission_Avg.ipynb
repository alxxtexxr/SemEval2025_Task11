```python
import os
import pandas as pd
from datasets import load_dataset
```

```python
# dirs = os.listdir('test')
# dirs = [
#     'track_a_202501150828', # BEST anger, joy, sadness (need improvement: disgust, fear, surprise)
#     'track_a_63_3', # BEST disgust
# ]
base_df = pd.read_csv('test/track_a_202501150828/pred_sun.csv')
disgust_df = pd.read_csv('test/track_a_63_3/pred_sun.csv')
fear_df = pd.read_csv('test/track_a_45/pred_sun.csv')
surprise_df = pd.read_csv('test/track_a_42_1/pred_sun.csv')

# dfs = [pd.read_csv(os.path.join('test', dir, 'pred_sun.csv')) for dir in dirs]
id_df = base_df[['id']]
# dfs = [df.drop('id', axis=1) for df in dfs]
```

```python
disgust_df.loc[disgust_df['disgust'] != 1, :] = 0
fear_df.loc[fear_df['fear'] != 1, :] = 0
surprise_df.loc[surprise_df['surprise'] != 1, :] = 0
```

```python
dfs = [base_df, disgust_df, fear_df, surprise_df]
dfs = [df.drop('id', axis=1) for df in dfs]
```

```python
avg_df = sum(dfs)/len(dfs)
avg_df
```

```python
one_hot_df = (avg_df >= 0.5).astype(int)
one_hot_df
```
