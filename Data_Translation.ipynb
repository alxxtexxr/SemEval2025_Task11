```python
import json
import pandas as pd
from openai import OpenAI

client = OpenAI(api_key=json.load(open('secrets.json'))['OPENAPI_SECRET_KEY'])
```

```python
data_path = 'go_emotions_fear_single_12_comb_26.csv'
```

```python
df = pd.read_csv(data_path)
df.head()
```

```python
# Function to translate text using OpenAI
def translate_text(text, target_lang="Sunda"):
    completion = client.chat.completions.create(
        model="gpt-4o",
        messages=[
            # {"role": "system", "content": "You are a helpful assistant."},
            {
                "role": "user",
                "content": f"Terjemahkan teks ini dari bahasa Inggris ke bahasa {target_lang}: {text}",
            },
        ],
        max_tokens=100,
    )
    return completion.choices[0].message.content

# Translate each text
for i, row in df.iterrows():
    print(f"# {i+1}")
    print(row['text'])
    translated_text = translate_text(row['text'], target_lang="Sunda")
    print(translated_text)
    df.at[i, 'translated_text'] = translated_text
    print()
```

```python
save_path = f'{data_path.split('.')[-0]}_translated.csv'
df.to_csv(save_path, index=False)
print("Saved to:", save_path)
```
