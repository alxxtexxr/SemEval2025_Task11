```python
import pandas as pd
```

```python
data_path = 'go_emotions_fear_single_12_comb_26_translated_cleaned_curation.csv'
selected_emotion = 'fear'
```

```python
df = pd.read_csv(data_path)
print("DF size:", len(df))
df.head()
```

```python
if 'curation_status' in df.columns:
    df_cleaned = df[df['curation_status'] == 1]
    print("Cleaned DF size (after curation):", len(df_cleaned))
else:
    df_cleaned = df.copy()
```

```python
print("Num. of duplicates:", df_cleaned['text'].duplicated().sum())
df_cleaned = df_cleaned.drop_duplicates(subset=['text'], keep='first')
print("Num. of duplicates (after cleaning duplicates):", df_cleaned['text'].duplicated().sum())
print("Cleaned DF size (after cleaning duplicates):", len(df_cleaned))
```

```python
cond_r = df_cleaned['text_translated'].str.contains("r/")
cond_u = df_cleaned['text_translated'].str.contains("/u/")
cond_name = df_cleaned['text_translated'].str.contains("NAME]")
cond_maaf = df_cleaned['text_translated'].str.contains("Maaf")
# cond_terjemah = df_cleaned['text_translated'].str.contains('terjemah')
# cond_double = df_cleaned['text_translated'].str.contains(r'(.)\1{2,}')

df_cleaned = df_cleaned[~(cond_r | cond_u | cond_name)]
print("Cleaned DF size:", len(df_cleaned))
df_cleaned.head()
```

```python
def create_num_emotions_col(row):
    return int(sum(row[['anger', 'disgust', 'fear', 'joy', 'sadness', 'surprise']].tolist()))

df_cleaned.apply(create_num_emotions_col, axis=1)
df_cleaned.loc[:, 'num_emotions'] = df_cleaned.apply(create_num_emotions_col, axis=1)
df_cleaned.head()
```

```python
df_cleaned_single = df_cleaned[(df_cleaned[selected_emotion] == 1) & (df_cleaned['num_emotions'] == 1)]
df_cleaned_comb = df_cleaned[(df_cleaned[selected_emotion] == 1) & (df_cleaned['num_emotions'] > 1)]

print("Total cleaned data with single emotion:", len(df_cleaned_single))
print("Total cleaned data with combined emotions:", len(df_cleaned_comb))
```

```python
save_path = f'go_emotions_{selected_emotion}_single_{len(df_cleaned_single)}_comb_{len(df_cleaned_comb)}_translated_cleaned_curated.csv'
df_cleaned.to_csv(save_path, index=False)
print("Saved to:", save_path)
```
